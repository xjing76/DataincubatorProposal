Here are some sample img from this account , one of the easiest way to approach to this would simply compute the MSE of the two img matrix. 

```python
import numpy as np
def mse(imageA, imageB):
	l, w, c = imageA.shape
	l1, w2, c2 = imageB.shape
	imageA = cv2.resize(imageA, (min(l,l1), min(w,w2))) 
	imageB = cv2.resize(imageB, (min(l,l1), min(w,w2))) 
	err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
	err /= float(imageA.shape[0] * imageA.shape[1])
	return err
```
